```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
salary_df = pd.read_csv("Resources/Salary_Data.csv")
salary_df.head()
```

```python
# Splitting 'location' column into 4 columns (city, state, country, extra)
salary_df[['city', 'state', 'country', 'extra']] = salary_df.location.str.split(",", expand=True)
salary_df.head()
```

```python
# Unique names of 'country' column

salary_df.country.unique()
```

```python
# Unique names of 'extra' column

salary_df.extra.unique()
```

```python
# Unique names of 'country' columns 
# Deleted all names of countries except 'None' and 'United States'

array = [None, ' United States']

salary_df = salary_df.loc[salary_df.country.isin(array)]
salary_df.head()
```

```python
salary_df.state.unique()
```

```python
salary_df = salary_df[salary_df.state != ' Israel']
salary_df.state.unique()
```

```python
salary_df.drop(['location', 'country', 'extra'], axis=1, inplace=True)
```
